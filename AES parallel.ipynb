```python
pre_mat=np.array([[2,3,1,1],[1,2,3,1],[1,1,2,3],[3,1,1,2]])
```

```python
@njit()
def galoisMult(a, b):
    p = 0
    hiBitSet = 0
    for i in range(8):
        if b & 1 == 1:
            p ^= a
        hiBitSet = a & 0x80
        a <<= 1
        if hiBitSet == 0x80:
            a ^= 0x1b
        b >>= 1
    return p % 128
galoisMult(2,255)
npgaloisMult = np.frompyfunc(galoisMult, 2, 1)
```

```python
# @jit(nopython=False)
# @vectorize(["int32(int32)"])
# @njit(parallel=True)
def shift_rows(res):
    for i in range(1,res.shape[0]):
        res[i:i+1]=np.roll(res[i:i+1],-i,axis=1)
    return res.astype('int32')
```

```python
# @jit(nopython=False)
@vectorize(["int32(int32, int32)"])
def add_rnd_keys(x,y):
    return x^y

# @jit(nopython=False)
# @vectorize(["int32(int32, int32)"])
def custom_multiply(x, y):
    return np.array([np.bitwise_xor.reduce(npgaloisMult(row,column)) for row in x for column in y.T]).reshape(x.shape[0], y.shape[1])

# @jit(nopython=True)
def npmix_col(res):
    return custom_multiply(pre_mat,res)
```
